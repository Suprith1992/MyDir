## Drop rows with Null Values

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.tail(3)
```

```python
nba.dropna(how = "all", inplace = True) #arg how = "all" it drops a row if its all values are Nan, 
# by default it will be how = "any" it drops a row if even single field has nan value
```

```python
nba.tail(3)
```

```python
nba.head(3)
```

```python
# to delete rows only if particular row has nan values

nba.dropna(subset=['College','Salary'])
```

## .astype() method

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(6)
```

```python
nba.dtypes
```

```python
nba["Salary"] = nba["Salary"].astype("int")
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba.head(3)
```

## Rank Values with the `.rank()` Method

```python
nba = pd.read_csv("nba.csv")
nba.dropna(how="all").tail()
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head()
```

```python
nba.sort_values("Salary",ascending=False).head()
```

```python
nba["Salary Rank"] = nba["Salary"].rank().astype("int")
nba.head()
```
